```python
from os import path
import os
from pydub import AudioSegment
```

```python
broken = {}
```

```python
def turnMp3toWav(source, sampling, numChan, dest, neededData):
    for fileName in os.listdir(source):
        f = os.path.join(source, fileName)
        if os.path.isfile(f) and fileName in neededData:
#             print(source + dest+ fileName[:-4] + ".wav")
            try:
                sound = AudioSegment.from_mp3(f)
                sound.export(dest+ fileName[:-4] + ".wav", format="wav")
            except:
                neededData.pop(fileName)
                broken[fileName] = 1
            
    return neededData
```
