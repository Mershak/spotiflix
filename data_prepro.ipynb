```python
import pandas as pd

spotify_data = pd.read_csv("./echonest.csv", low_memory=False, header=[2])
folderDest = []
fileName = []
```

```python
for i in range(1, spotify_data.shape[0]):
    id = str(int(spotify_data["track_id"][i]))
    paddedId = '0'*(6-len(id)) + id

    folderDest.append('/' + paddedId[:3])
    fileName.append(paddedId + ".wav")
```

```python
processed = pd.DataFrame(list(zip(
                                 folderDest, fileName, 
                                 round(spotify_data["acousticness"][1:],1), round(spotify_data["danceability"][1:],1),
                                 round(spotify_data["energy"][1:],1), round(spotify_data["instrumentalness"][1:],1),
                                 round(spotify_data["liveness"][1:],1), round(spotify_data["speechiness"][1:],1),
                                 round(spotify_data["tempo"][1:]), round(spotify_data["valence"][1:],1)
                                 )
                             ),
                   columns =['folder', 'fileName', 'acousticness', "danceability", "energy", "instrumentalness",
                            "liveness", "speechiness", "tempo", "valence"])
processed
```

```python
processed.to_csv("spotify_data2.csv", index=False)
```
